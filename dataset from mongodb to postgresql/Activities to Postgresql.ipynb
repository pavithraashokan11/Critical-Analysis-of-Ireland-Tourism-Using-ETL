# Establishing connect to MongoDB

```python
import pymongo
import pandas as pd
from pymongo import MongoClient
client = MongoClient('mongodb://root:rootpassword@localhost:27017/')
db = client['IRL']

#Creating a new collection 
collection=db.Activities
act=pd.DataFrame(list(collection.find()))
act.head(10)
```

```python
act=act.drop(['_id','@context','image',], axis=1)
```

```python
act.head()
```

```python
Name = act['name'].tolist()
City = act['address'].apply(lambda x: x['addressLocality']).tolist()
County = act['address'].apply(lambda x: x['addressRegion']).tolist()
Country = act['address'].apply(lambda x: x['addressCountry']).tolist()
Latitudes = act['geo'].apply(lambda x: x['latitude']).tolist()
Longitudes = act['geo'].apply(lambda x: x['longitude']).tolist()
URL = act['url'].tolist()
Telephone= act['telephone'].tolist()
Tags= act['tags'].tolist()
type_of_place=pd.DataFrame(columns=['Type of Activity'])

for data in act["tags"]:
    if len(data) >= 2:
        Tags.append(data[1])
    else:
        Tags.append(data[0])
type_of_place['Type of Venue'] = Tags
```

```python
Tag = []
for i in range(len(Tags)):
    for j in range(len(Tags[i])):
        if j % 2 != 0:
            Tag.append(Tags[i][j])

print(Tag)
```

```python
Type= [] 
type_of=pd.DataFrame(columns=['Type of Venue'])

for data in act["@type"]:
    if len(data) >= 2:
        Type.append(data[1])
    else:
        Type.append(data[0])
type_of['Type of Venue'] = Type
```
